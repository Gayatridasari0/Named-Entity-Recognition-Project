## Phase 3

### Accuracy Assesment

```python
def retrieve_entity_positions(sentence, entity_text):
    """Find the start and end positions of the entity in the sentence."""
    start = sentence.find(entity_text)
    if start == -1:  # If the entity is not found, then return None
        return None
    end = start + len(entity_text)
    return start, end
```

## Converting Spacy Structure according to NER

```python
def format_conversion_from_spacy_to_manual(spacy_NER):
    """Convert spaCy format (word, tag) to manual format (start, end, tag)."""
    annotations_after_conversion = []

    for entry in spacy_NER:
        sentence = entry['sentence']
        entities = []

        for entity_text, entity_tag in entry['entities']:
            positions = retrieve_entity_positions(sentence, entity_text) #get each entity start & end index positions just like manual JSON output
            if positions:
                start, end = positions
                entities.append([start, end, entity_tag])

        annotations_after_conversion.append({'sentence': sentence, 'entities': entities})

    return annotations_after_conversion
```

```python
true_positives = defaultdict(int)
false_positives = defaultdict(int)
false_negatives = defaultdict(int)
false_positives_list = []
false_negatives_list = []
def calculate_metrics(manual_NER, spacy_NER):

    for manual, spacy in zip(manual_NER, spacy_NER):
        # Convert manual annotations into a set of tuples (start, end, label)
        set_manual_entities = set([tuple(entity) for entity in manual['entities']])

        # Convert spaCy annotations into a set of tuples (start, end, label)
        set_spacy_entities = set([tuple(entity) for entity in spacy['entities']])
            
        for entity in set_spacy_entities:
            if entity in set_manual_entities:
                true_positives[entity[2]] += 1
            else:
                false_positives[entity[2]] += 1
                false_positives_list.append(entity)
            
        for entity in set_manual_entities:
            if entity not in set_spacy_entities:
                false_negatives[entity[2]] += 1
                false_negatives_list.append(entity)
```

```python
# Convert spaCy format to manual format
spacy_after_conversion = format_conversion_from_spacy_to_manual(spacy_output)

# Now calculate TP, FP, FN using the converted spaCy annotations
calculate_metrics(Manual_NER_list, spacy_after_conversion)
```

```python
print("------------------------------------------------------------------------")
print(f"total_tp: ",true_positives)
print(f"total_fp: ", false_positives)
print(f"total_fn: ",false_negatives)
print("------------------------------------------------------------------------")
print("fp list",false_positives_list)
print("------------------------------------------------------------------------")
print("fn list",false_negatives_list)
print("------------------------------------------------------------------------")
```

```python
labels = set(true_positives.keys()).union(false_positives.keys()).union(false_negatives.keys())

for label in labels:
    tp = true_positives[label]
    fp = false_positives[label]
    fn = false_negatives[label]
    
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    print("------------------------------------------------------------------------")
    print(f"Tag: {label}")
    print(f"  Precision: {precision:.2f}")
    print(f"  Recall:    {recall:.2f}")
    print(f"  F1 Score:  {f1:.2f}\n")
```

The performance of the spaCy NER pipeline on different entity categories is varied. Entities like "GPE" (Geopolitical Entities) and "NORP" (Nationalities or Religious/Political Groups) have relatively higher F1 scores of 0.69 and 0.72, respectively, showing that for these entities, precision and recall are better balanced in identifying them. In contrast, categories like "LOC" depict the lowest F1 score at 0.21, while "MONEY" follows at 0.13, showing poor detection and classification in those categories. For "PERSON" and "ORG", it showed a medium performance of 0.43 and 0.29 F1 scores, respectively. These variations underline the model's differing efficacy across entity types, probably because of factors such as the diversity and representation of entities in the training data or due to the inherent complexity of certain categories. Thus, this allows for targeted data augmentation and fine-tuning of the model to improve performance, particularly for the underperforming categories.
